```python
import pandas as pd

df1_new = pd.read_excel(r'/Users/kimda/Documents/장애인고용/company_pre.xlsx')
df2 = pd.read_excel(r'/Users/kimda/Documents/장애인고용/data_clusters.xlsx')
```

```python
df2['희망지역'].unique()
```

```python
#comAddr값 띄어쓰기 이후 삭제
df1_new['New_Address'] = df1_new['compAddr'].apply(lambda x: x.split()[0])
df1_new['New_Address'].unique()
```

```python
# 제주, 기타, 전남, 전체, = 데이터에 없는거라 빈리스트반환
# 경상: 경상남도,경상북도 / 전라: 전라남도,전라북도
```

```python
# '희망지역'열의 글자가 'New_Address'의 값에 포함되는지 확인하는 함수

def check_inclusion(word, df1_new):
    result = []
    for value in df1_new['New_Address']:
        if all(char in value for char in word):
            result.append(value)
    return list(set(result))

# df2에 새로운 열 추가
df2['char'] = df2['희망지역'].apply(lambda x: check_inclusion(x, df1_new))

df2[11438:11445]
```

```python
import pandas as pd

# 매칭할 열 이름 설정
column_1 = "char"
column_2 = "New_Address"

# 새로운 열에 추가될 데이터를 담을 리스트 초기화
new_column_values = []

# df2의 각 행에 대해 순회
for index, row in df2.iterrows():
    # 매칭할 열1의 값 (리스트)
    value_list = row[column_1]
    
    # 필터링된 행의 'busplaName' 값을 담을 리스트 초기화
    matching_b_values = []

    # df1_new의 각 행에 대해 순회
    for _, sub_row in df1_new.iterrows():
        # 매칭할 열2의 값 (단일 값)
        address_value = sub_row[column_2]
        
        # value_list에 address_value가 있는지 확인
        if address_value in value_list:
            matching_b_values.append(sub_row['busplaName'])
    
    # 결과를 리스트에 추가
    new_column_values.append(matching_b_values)

# 새로운 열 생성
df2['MatchingBusplaName3'] = new_column_values



# 결과 출력
df2
```
