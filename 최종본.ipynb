#### 희망직종 텍스트 데이터 클러스터링

```python
# 희망직종과 jobNm을 합친 후 클러스터링
df2_path = 'data_clusters.xlsx'
company_path = 'company_pre.xlsx'

data = pd.read_excel(df2_path)
df = pd.read_excel(company_path)


all_jobs = pd.concat([data['희망직종'], df['jobNm']])
# company_df==company의 regagnName 열 전처리, 음료조리사->바리스타로 변경
data['희망직종'] = data['희망직종'].replace('음료 조리사', '바리스타')
# TF-IDF 벡터화
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(all_jobs)

# K-Means 클러스터링
num_clusters = 50  # 적절한 클러스터 수를 설정
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(tfidf_matrix)

# 클러스터 라벨 추가
data['사람직종클러스터'] = kmeans.predict(tfidf_vectorizer.transform(data['희망직종']))# 장애인
df['회사직종클러스터'] = kmeans.predict(tfidf_vectorizer.transform(df['jobNm']))# company

# 클러스터링 결과 확인
df.head(), data.head()
df.to_excel("company_test6.xlsx",index=False)
data.to_excel("people_test6.xlsx",index=False)
```

```python
#comAddr값 띄어쓰기 이후 삭제
df2['New_Address'] = df2['compAddr'].apply(lambda x: x.split()[0])
df2['New_Address'].unique()
# 제주, 기타, 전남, 전체, = 데이터에 없는거라 빈리스트반환
# 경상: 경상남도,경상북도 / 전라: 전라남도,전라북도
```

```python
# '희망지역'열의 글자가 'New_Address'의 값에 포함되는지 확인하는 함수

def check_inclusion(word, df2):
    result = []
    for value in df2['New_Address']:
        if all(char in value for char in word):
            result.append(value)
    return list(set(result))

# df2에 새로운 열 추가
df2['char'] = df2['희망지역'].apply(lambda x: check_inclusion(x, df2))
```

```python
import pandas as pd

# 매칭할 열 이름 설정
column_1 = "char"
column_2 = "New_Address"

# 새로운 열에 추가될 데이터를 담을 리스트 초기화
new_column_values = []

# df2의 각 행에 대해 순회
for index, row in df2.iterrows():
    # 매칭할 열1의 값 (리스트)
    value_list = row[column_1]
    
    # 필터링된 행의 'busplaName' 값을 담을 리스트 초기화
    matching_b_values = []

    # df1_new의 각 행에 대해 순회
    for _, sub_row in df2.iterrows():
        # 매칭할 열2의 값 (단일 값)
        address_value = sub_row[column_2]
        
        # value_list에 address_value가 있는지 확인
        if address_value in value_list:
            matching_b_values.append(sub_row['busplaName'])
    
    # 결과를 리스트에 추가
    new_column_values.append(matching_b_values)

# 새로운 열 생성
df2['MatchingBusplaName3'] = new_column_values



df2.to_excel("where_data_cluster.xlsx")

### "희망지역',busplaName,compAddr삭제 ###
```

```python
# '희망지역'열과 'New_Address'열의 지역이 일치하지 않을 시 삭제. 확인용으로 연번 18 확인바람

def filter_dataframe(df2):
    # '희망지역' 열의 값이 'New_Address' 열의 값에 포함되는지 확인하여 boolean mask 생성
    mask = df2.apply(lambda row: row['희망지역'] in row['New_Address'] or all(char in row['New_Address'] for char in row['희망지역']), axis=1)
    
    # boolean mask를 사용하여 데이터프레임 필터링
    filtered_df = df2[mask]
    
    return filtered_df


# 함수 호출
filtered_df = filter_dataframe(df2)


filtered_df.to_excel('전처리최종본.xlsx', index=False)

# 결과 출력
filtered_df
```
